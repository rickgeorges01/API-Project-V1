```python
#import des librairies
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier

#Ajout de la dataset
iris = pd.read_csv("./sample_data/iris.csv")

x = iris.loc[:, "petal_length"]
y = iris.loc[:, "petal_width"]
lab = iris.loc[:, "species"]

#valeurs
longueur = 2.5
largeur = 0.75
k = 5  # Modification a été faites ici

#Configuration du graphique
plt.axis('equal')
plt.scatter(x[lab == 0], y[lab == 0], color ='g', label = 'setosa')
plt.scatter(x[lab == 1], y[lab == 1], color = 'r', label = 'versicolor')
plt.scatter(x[lab == 2], y[lab == 2], color = 'b', label = 'virginica')
plt.scatter(longueur, largeur, color='k')
plt.legend()

#algo knn
d = list(zip(x, y))
model = KNeighborsClassifier(n_neighbors=k)
model.fit(d, lab)
prediction = model.predict([[longueur, largeur]])
# Afficher la prédiction
print("Prédiction:", prediction)

#Affichage résultats
txt = "Résultat : "
if prediction[0] == 0: txt += "setosa"
if prediction[0] == 1: txt += "versicolor"
if prediction[0] == 2: txt += "virginica"
plt.text(3, 0.5, f"largeur : {largeur} cm longueur :{longueur} cm", fontsize=12)
plt.text(3, 0.3, f"k : {k}", fontsize=12)
plt.text(3, 0.1, txt, fontsize=12)
plt.show()

#OBSERVATION
#Lorsque l'on augmente le nombre de voisins dans l'algorithme KNN (en passant de k = 3 à k = 5), la prédiction pour un point donné est influencée par un plus grand nombre de voisins.
#Cela peut rendre l'algorithme plus robuste aux données aberrantes, car la prédiction n'est plus autant influencée par les voisins les plus proches.
#Cependant, cela peut également rendre l'algorithme moins sensible aux subtiles variations de données,
#ce qui peut parfois entraîner des erreurs de classification, surtout si les classes sont très entremêlées.
#L'effet spécifique de cette modification dépend de la distribution des données dans l'espace des caractéristiques.
#Dans le cas de l'Iris dataset, augmenter k pourrait soit stabiliser la prédiction en réduisant le bruit,
#soit rendre la prédiction moins précise si les classes ne sont pas bien séparées à l'échelle de k.
```
